# Statistics


1. What is hypothesis testing in statistics?
   - Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population based on sample data. It involves formulating two competing statements — a null hypothesis (H₀) and an alternative hypothesis (H₁) — and then using sample data and probability theory to determine whether there is enough evidence to reject the null hypothesis.
2. What is the null hypothesis, and how does it differ from the alternative
hypothesis?
   - The null hypothesis (H₀) is a statement that there is no effect, no difference, or no relationship in the population. It represents the default assumption.

The alternative hypothesis (H₁) is the statement that contradicts the null hypothesis, suggesting that there is a significant effect, difference, or relationship.

In short:

 - Null Hypothesis → Assumes no change or effect

 - Alternative Hypothesis → Assumes some change  or effect exists
 3. Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.
    - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It is commonly set at 0.05 or 0.01.

If the p-value ≤ α → Reject H₀
If the p-value > α → Fail to reject H₀

It controls the risk of making a Type I error and helps determine whether the observed data is statistically significant.

4. What are Type I and Type II errors? Give examples of each.
  - Type I Error: Rejecting the null hypothesis when it is true.
Example: Saying a new drug works when it actually does not.

   - Type II Error: Failing to reject the null hypothesis when it is false.
Example: Saying a new drug does not work when it actually does.

Type I Error = False Positive
Type II Error = False Negative

5. What is the difference between a Z-test and a T-test? Explain when to use
each.
   - | Z-test                                 | T-test                               |
| -------------------------------------- | ------------------------------------ |
| Used when population variance is known | Used when variance is unknown        |
| Large sample size (n > 30)             | Small sample size (n ≤ 30)           |
| Normal distribution assumed            | Used when population SD is not given |
### use -
 - Use Z-test when the sample size is large and population variance is known.

-  Use T-test when the sample size is small and population variance is unknown.

```python
#Q7. Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#                        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#                        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
#                       50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

pop_mean = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

z = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))
p_value = 1 - stats.norm.cdf(abs(z))

print("Z-score:", z)
print("P-value:", p_value)
```

```python
# Q8. Write a Python script to simulate data from a normal distribution and
# calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

data = np.random.normal(100, 10, 1000)

mean = np.mean(data)
std = np.std(data)
n = len(data)

ci = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(n))
print("95% Confidence Interval:", ci)

plt.hist(data)
plt.title("Normal Distribution")
plt.show()
```

```python
# Q9. Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt

data = [10, 12, 15, 18, 20, 22, 25]

mean = np.mean(data)
std = np.std(data)

z_scores = [(x - mean) / std for x in data]
print("Z-scores:", z_scores)

plt.hist(z_scores)
plt.title("Z-score Histogram")
plt.show()
```

 Explain: what the Z-scores represent
in terms of standard deviations from the mean.
   - Z-score tells how far a value is from the mean in terms of standard deviations.

     Z = (x - mean) / standard deviation



- Z = 0 → Value is exactly at the mean

- Z > 0 → Value is above the mean

- Z < 0 → Value is below the mean

- |Z| = 2 → Value is 2 standard deviations away from mean

The histogram shows the standardized data centered around 0, helping identify how spread out the values are relative to the mean.
